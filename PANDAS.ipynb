```python
import csv
csv_columns = ['name', 'age','address']
dict = [
    {'name':'sameer','age':45,'address':'kathmandu'},
    {'name':'ameer','age':5,'address':'thandu'}
    ]

csvfile = open("data.csv",'w')
writer = csv.DictWriter(csvfile, fieldnames = csv_columns)

writer.writeheader()
writer.writerows(dict)
csvfile.close()


```

```python
import pandas as pd
df1 = pd.DataFrame({'name':['sameer','ram','shyam','bhmimin'],
        'age' :[23,45,67,34],
        'add' :['dhapakhel', 'lalitpur','patan','toti']
       })

df2 = pd.DataFrame({'name':['sameer','ram','shyam','bhmimin'],
        'age' :[23,45,67,34],
        'add' :['dhapakhel', 'lalitpur','patan','toti']
       })
df = pd.concat([df1,df2]).reset_index()
df = df[['name','age','add']]
df.to_json('concat.jason')
df
```
