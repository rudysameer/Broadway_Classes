```python
# function with argrument and no return type

def cal(x,y,z):#parameter
  
    a = x*y       #local variable
    v = x*y*z       #local variable
    print("the area is", a)
    print("the volume is", v)
    
    
l = int(input("enter length :"))           #global variable
b = int(input("enter breadth :"))          #global variable
h = int(input("enter height "))             #global variable    
cal(l,b,h)    #l,b,h argument
```

```python
# Filter,Map
```

```python
a =[1,2,3,4,5,6,9,7,11]
b=list(filter(lambda x : x%2==0,a))
print(b)
```

```python
a = ["ronaldinho","MaRADONA","MESSI","GEROREIO","adriano","manueal neuer"]
b= list(filter(lambda x:x[0].isupper(),a))
print(b)
```

```python
# MAP FUNCTION
a = ["ronaldinho","MaRADONA","MESSI","GEROREIO","adriano","manueal neuer"]
b=list(map(lambda x : x.lower(),a))
print(b)
```

```python
a =[1,2,4,5,7,8,9,0,23,32]
b=list(map(lambda x: x*4,a))
print(b)
```

```python
a=list()
for i in range(1,11):
    
    a.append(i)
    
print(a)
b = list(map(lambda x:x*2,a))
print(b)
```

```python
# WAP to create basic calculator using return type function
def calculator(a,b):
    add=a+b
    sub=a-b
    multiply=a*b
    divide=a/b
    return add,sub,multiply,divide



a = int(input("enter the number a : "))
b = int(input("enter the number b : "))
x=calculator(a,b)
a,s,m,d = x

print("the addition of the numbers are :",a)
print("the subtraction of the numbers are :",s)
print("the multiplication of the numbers are :",m)
print("the division of the numbers are :",d)

```
